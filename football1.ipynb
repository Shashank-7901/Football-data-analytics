```python
#importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/drive/MyDrive/football data analysis/EPL_20_21.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#create two columns

df['MinsPerMatch'] = (df['Mins'] / df['Matches']).astype(int)
df['GoalsPerMatch'] = (df['Goals'] / df['Matches']).astype(float)
df.head()
```

```python
#total goals

totGoals = df['Goals'].sum()
print(totGoals)
```

```python
#penalty goals

totPenaltyGoALS = df['Penalty_Goals'].sum()
print(totPenaltyGoALS)
```

```python
#attempted penalty goals

attPenaltyGoals = df['Penalty_Attempted'].sum()
print(attPenaltyGoals)
```

```python
#piechart for penalties scored vs missed

plt.figure(figsize =(6,13))
pl_notScored = df['Penalty_Attempted'].sum()-totPenaltyGoALS
data=[pl_notScored,totPenaltyGoALS]
labels=['Penalty Missed','Penalty Scored']
color=sns.color_palette('Set2')
plt.pie(data,labels=labels,colors=color,autopct='%.0f%%')
plt.show()
```

```python
#unique positions

df['Position'].unique()
```

```python
#forward players

df[df['Position'] == 'FW']
```

```python
#players from different nations

np.size((df['Nationality'].unique()))
```

```python
#most players from which country

nation = df.groupby('Nationality').size().sort_values(ascending=False)
nation.head().plot(kind='bar',figsize=(12,6),color = sns.color_palette("magma"))
```

```python
#clubs with maximum players in their squad

df['Club'].value_counts().nlargest(10).plot(kind='bar',color = sns.color_palette("viridis"))
```

```python
#clubs with minimum players in their squad

df['Club'].value_counts().nsmallest(10).plot(kind='bar',color = sns.color_palette("viridis"))
```

```python
#players based on age groups

u20 = df[df['Age']<=20]
age20_25 = df[(df['Age']>20) & (df['Age']<=25)] 
age25_30 = df[(df['Age']>25) & (df['Age']<=30)]
age30 = df[df['Age']>30] 
```

```python
x=np.array([u20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),age30['Name'].count()])
mylabels = ["<=20",">20&<=25",">25&<=30",">30"]
plt.title('Total players with age',fontsize=25)
plt.pie(x, labels=mylabels,autopct="%.1f%%")
plt.show()
```

```python
#total under 20 players in each club

players_under_20 = df[df['Age']<20]
players_under_20['Club'].value_counts().plot(kind='bar',color=sns.color_palette("cubehelix"))
```

```python
#average age of players in each club

plt.figure(figsize=(12,6))
sns.boxplot(x='Club',y='Age',data=df)
plt.xticks(rotation=90)
```

```python
avg_age = df.groupby('Club').size()
data = (df.groupby('Club')['Age'].sum())/avg_age
data.sort_values(ascending=True)
```

```python
#most goals scored by players

top_goal_scoreres = df[['Name','Club','Goals','Matches']].nlargest(n=10,columns = 'Goals')
top_goal_scoreres
```

```python
#average goals scored per match

avg_goal_score = df[['Name','Club','Goals','GoalsPerMatch','Matches']].nlargest(n=10,columns = 'GoalsPerMatch')
avg_goal_score
```
